```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
#loading CSV file
df=pd.read_csv('dataset.csv')
df
```

```python
#Check for duplicates
x=df.duplicated()
print("Duplicated :",x.sum())

df=df.drop(columns=['m_dep','px_height','px_width','mobile_wt','sc_w','sc_h','talk_time','n_cores'])
df
```

```python
#Check for null values

x=df.notnull().sum()
print(x)

#drop null values


df.dropna()
```

```python
#Defining class names
class_names={0 : 'low cost',
1 : 'medium cost',
2 : 'high cost',
3 : 'very high cost'}

#Splitting the dataset
ds=df.to_numpy()
X=ds[:,:-1]
y=ds[:,-1]

from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15,shuffle=True)

df
```

```python
#Model Training
model=KNeighborsClassifier(n_neighbors=4)
model.fit(X_train,y_train)
```

```python
print(classification_report(y_true=y_test,y_pred=model.predict(X_test)))
```
