```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor
import pandas as pd
```

```python
#loading CSV file
df=pd.read_csv('dataset.csv')

df
```

```python
#Check for duplicates
x=df.duplicated()
print("Duplicated :",x.sum())

#drop duplicates

df=df.drop_duplicates()

print(df.duplicated().sum())


#drop unnecessary columns
df=df.drop(df.columns[[0,1,3,4,6]],axis=1)
df
```

```python
#Check for null values

x=df.notnull().sum()
print(x)

#drop null values

df=df.dropna()

df=df.drop(df.columns[[9,10]],axis=1)
df
```

```python
from sklearn.preprocessing import LabelEncoder

lb_body=LabelEncoder()
lb_drivetrain=LabelEncoder()
lb_fuel=LabelEncoder()
lb_make=LabelEncoder()

df=df.drop(df.columns[[5,6,8]],axis=1)

df['body']=lb_body.fit_transform(df['body'])
df['drivetrain']=lb_drivetrain.fit_transform(df['drivetrain'])
df['fuel']=lb_fuel.fit_transform(df['fuel'])
df['make']=lb_make.fit_transform(df['make'])

print(lb_body.classes_,"\n",lb_drivetrain.classes_,"\n",lb_fuel.classes_,"\n",lb_make.classes_,"\n")

df
```

```python
#Splitting the dataset
ds=df.to_numpy()

y=ds[:,1]

df=df.drop(df.columns[[1]],axis=1)
ds=df.to_numpy()

X=ds


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0,shuffle=True)

y.shape
```

```python
#Model Training
model=RandomForestRegressor()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

y_pred=model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred) 
print(f'Mean Squared Error: {mse}') 
print(f'R-squared: {r2}')

import pickle

with open('price_model.pkl', 'wb') as file: pickle.dump(model, file)
```
